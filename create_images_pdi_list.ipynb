```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

## (1) Create the image of a paraboloid with one axis scaled (like an oval paraboloid).

```python
x = np.linspace(-10, 10, 100, endpoint=True)
y = np.linspace(-10, 10, 100, endpoint=True)
X, Y = np.meshgrid(x, y)
Z = X ** 2 + 2*Y ** 2 # function aplied

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Paraboloid One-Axis-scaled')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

## (2) Create the image of a rotated sin using rotation of coordinates.

```python
x = np.linspace(-10, 10, 100, endpoint=True)
y = np.sin(x)
angle = np.pi / 4

x_rotated = x * np.cos(angle) + y * np.sin(angle)  # x'
y_rotated = -x * np.sin(angle) + y * np.cos(angle) # y'

plt.title('Rotated Sine Function')
plt.plot(x_rotated, y_rotated)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

## (3) Create the image of a gaussian.

```python
x = np.linspace(-10, 10, 100, endpoint=True)
y = np.exp(-((x - 0) ** 2) / (2 * 1 ** 2)) / (1 * np.sqrt(2 * np.pi)) # std = 1; mean = 0

plt.title('Gaussian')
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

plt.show()
```

## (4) Create a function that generates the image of a Gaussian optionally rotated by an angle \theta and with mx, my, sx, sy as input arguments.

```python
mx = float(input("Valor de mx: ")) # translate in x-axis
my = float(input("Valor de my: ")) # translate in y-axis
sx = float(input("Valor de sx: ")) # scale in x-axis
sy = float(input("Valor de sy: ")) # scale in y-axis

angle = np.pi / 4

x = np.linspace(-5, 5, 100, endpoint=True)
y = np.exp(-((x - 0) ** 2) / (2 * 1 ** 2)) / (1 * np.sqrt(2 * np.pi)) # std = 1; mean = 0

x_rotated = x * np.cos(angle) + y * np.sin(angle)  # x'
y_rotated = -x * np.sin(angle) + y * np.cos(angle) # y'

x_transformed = (x_rotated * sx) + mx
y_transformed = (y_rotated * sy) + my

plt.title('Gaussian Rotated')
plt.plot(x_transformed, y_transformed)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

plt.show()
```
